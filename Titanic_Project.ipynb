# Age Imputation Function:
 Creating The function to fill the missing cells in the columns

```python
def fill_age(cols):
  Age=cols[0]
  Pclass=cols[1]
  if pd.isnull(Age):
    if Pclass==1:
      return 37
    elif Pclass==2:
      return 29
    else:
      return 24
  return Age
```

Empty cells in Age column is filled with the average ages based on the Passenger class(Pclass)

```python
df[['Age','Pclass']].apply(fill_age,axis=1)
```

```python
df['Age']=df[['Age','Pclass']].apply(fill_age,axis=1)
```

```python
sns.heatmap(df.isnull())
```

# Cabin Removal:
 Removing the Cabin column as it contains more missing values

```python
df.drop('Cabin',axis=1,inplace=True)
```

Removing the missing data in a row of Cabin column

```python
df.dropna(inplace=True)
```

#Feature Engineering
Machine can't understand the catagorical data like gender so we need to convert this into binary form male=1,female=0

```python
pd.get_dummies(df['Sex'])
# pd.get_dummies(df['Embarked'])
```

# Creating new columns with dummies

*   Categorical Encoding (Sex, Embarked):

```python
sex=pd.get_dummies(df['Sex'],drop_first=True)
```

```python
embarked=pd.get_dummies(df['Embarked'],drop_first=True)
```

```python
df.columns
```

# Dropping Redundant Columns:

```python
df.drop(['Sex','Embarked','Name','Ticket','PassengerId'],axis=1,inplace=True)
```

# The refined datasets

```python
df.head()
```

# Joining the different dataframes

```python
df=pd.concat([df,sex,embarked],axis=1)
```

# Model Building and Evaluation



```python
from sklearn.model_selection import train_test_split
```



*   Data Splitting:



```python
 X_train,X_test,y_train,y_test  =  train_test_split(df.drop('Survived',axis=1),df['Survived'],test_size=0.30,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```



*   Logistic Regression:



```python
log=LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
predictions=log.predict(X_test)
```

# Evaluation

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
